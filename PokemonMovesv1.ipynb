```python
td = tbody.find_all('td')
```

```python
name = []


for n in td[::8]:
    
   
    s = str(n.find('a').string)
    name.append(s)
    
    
print(name)
```

```python
cat = []


for c in td[2::8]:
    if (c.find('img')==None):
        cat.append(None)
    else:
        img = c.find('img')
        cat.append(img.get('alt'))
        
    
print(len(cat))
```

```python
moveType = []

for t in td[1::8]:
    moveType.append(str(t.find('a').string))
    
print(len(moveType))
```

```python
category = []
power = []
acc = []
PP = []
effect = []
prob = []

for c in td[2::8]:
    if (c.find('img')==None):
        category.append(None)
    else:
        img = c.find('img')
        category.append(img.get('alt'))

for p in td[3::8]:
    power.append(p.string)
for a in td[4::8]:
    acc.append(a.string)
    
for P in td[5::8]:
    PP.append(p.string)
for e in td[6::8]:
    effect.append(e.string)
for probVal in td[7::8]:
    prob.append(probVal.string)
    
Data = {'Move Name': name, 'Type': moveType, 'Category': category, 'Power': power, 'Accuracy': acc, 'Power Points': PP,
       'Effect' : effect, 'Probability': prob}

df = pd.DataFrame(Data)

display(df)
```

```python
drop_word1 = "Z-Move"
drop_word2 = "Dynamax move"
drop_word3 = "G-Max Move"


mask = df['Effect'].str.contains(drop_word1,case = False).fillna(False)

df = df[~mask]


mask2 = df['Effect'].str.contains(drop_word2,case = False).fillna(False)

df = df[~mask2]


mask3 = df['Effect'].str.contains(drop_word3,case = False).fillna(False)

df = df[~mask3]

df = df.reset_index()
df = df.drop(columns='index')

print(df.columns)
display(df)
```
